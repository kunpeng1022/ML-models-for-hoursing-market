```python
#Kunpeng Liu   kl3070
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer, make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import validation_curve
import warnings
warnings.filterwarnings('ignore')
df = pd.read_excel('AmesHousing.xls',header=0)
df=df.loc[:, df.columns != 'Order']
df=df.loc[:, df.columns != 'PID']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(df.loc[:, df.columns != 'SalePrice'],df['SalePrice'])

R_2={}
top_3=[]
for var in Cat_V:
    feature=pd.get_dummies(X_train[var])
    lr = LinearRegression().fit(feature,y_train)
    scores = cross_val_score(lr, feature, y_train, cv=5)
    R_2[var]=np.mean(scores)
    
R_2=sorted(R_2.items(), key=lambda x: x[1],reverse=True)
for x in range (0,3):
    top_3.append(R_2[x][0])
    
fig = plt.figure()
plt.scatter(df[top_3[0]],df['SalePrice']) 
plt.title(top_3[0] +' and SalePrice')
plt.xlabel(top_3[0])
plt.ylabel('SalePrice')
```

```python
#Task1.4
```

```python
X_train['MS SubClass']=X_train['MS SubClass'].astype(object)
X_train['Overall Qual']=X_train['Overall Qual'].astype(object)
X_train['Overall Cond']=X_train['Overall Cond'].astype(object)
```

```python
categorical= X_train.dtypes==object

imputer=make_pipeline(SimpleImputer(missing_values=np.nan,strategy='mean'))
preprocess = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), categorical),
    (imputer,~categorical)
)

print("Without StandardScaler:")
linear = make_pipeline(preprocess, LinearRegression())
valid_scores = cross_val_score(linear, X_train, y_train,cv=5)
print('LinearRegression mean cv R2: {:.4f}'.format(np.mean(valid_scores)))


ridge = make_pipeline(preprocess, Ridge())
valid_scores = cross_val_score(ridge, X_train, y_train,cv=5)
print('Ridge mean cv R2: {:.4f}'.format(np.mean(valid_scores)))

lasso = make_pipeline(preprocess, Lasso())
valid_scores = cross_val_score(lasso, X_train, y_train,cv=5)
print('Lasso mean cv R2: {:.4f}'.format(np.mean(valid_scores)))

from sklearn.linear_model import ElasticNet
ElasticNet = make_pipeline(preprocess, ElasticNet())
valid_scores = cross_val_score(ElasticNet, X_train, y_train,cv=5)
print('ElasticNet mean cv R2: {:.4f}'.format(np.mean(valid_scores)))
```

```python
imputer=make_pipeline(SimpleImputer(missing_values=np.nan,strategy='mean'),StandardScaler())

preprocess = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), categorical),
    (imputer,~categorical),
)

print("With StandardScaler:")
linear = make_pipeline(preprocess, LinearRegression())
valid_scores = cross_val_score(linear, X_train, y_train,cv=5)
print('LinearRegression mean cv R2: {:.4f}'.format(np.mean(valid_scores)))


ridge = make_pipeline(preprocess, Ridge())
valid_scores = cross_val_score(ridge, X_train, y_train,cv=5)
print('Ridge mean cv R2: {:.4f}'.format(np.mean(valid_scores)))

lasso = make_pipeline(preprocess, Lasso())
valid_scores = cross_val_score(lasso, X_train, y_train,cv=5)
print('Lasso mean cv R2: {:.4f}'.format(np.mean(valid_scores)))

from sklearn.linear_model import ElasticNet
ElasticNet = make_pipeline(preprocess, ElasticNet())
valid_scores = cross_val_score(ElasticNet, X_train, y_train,cv=5)
print('ElasticNet mean cv R2: {:.4f}'.format(np.mean(valid_scores)))

print("I think it helps Ridge regression and ElasticNet, but not too much for ohters")
```

```python
from sklearn.linear_model import ElasticNet

preprocess = make_column_transformer((OneHotEncoder(), Cat_V),(imputer,~categorical))
test = preprocess.fit_transform(X_train)

para={'l1_ratio':[.1,0.3,.5,0.7,.9,1]}
grid = GridSearchCV(ElasticNet(), para, cv=5)
grid.fit(test, y_train)

para = [.1,0.3,.5,0.7,.9,1]
plt.figure(1, figsize=(8, 6))

plt.plot(para, grid.cv_results_['mean_test_score'],label='cross-validation score')
plt.plot(para, grid.cv_results_['mean_train_score'],label='training score')

plt.ylabel('Scores')
plt.xlabel('Parameter')
lgd = plt.legend(loc=9,bbox_to_anchor=(0.8,1))
plt.show()


print(grid.best_params_)
print('ElasticNet mean cv R2: {:.4f}'.format(grid.cv_results_['mean_test_score'][para.index(grid.best_params_['l1_ratio'])]))
print("The result improves")
```

```python
para= {'alpha': [1e-3,1e-2, 1, 5, 10,30,50,70,100]}
grid = GridSearchCV(Lasso(), para, cv=3)
grid.fit(test, y_train)


para = [1e-3,1e-2, 1, 5, 10,30,50,70,100]

    
plt.figure(1, figsize=(8, 6))

plt.plot(para, grid.cv_results_['mean_test_score'],label='cross-validation score')
plt.plot(para, grid.cv_results_['mean_train_score'],label='training score')

plt.ylabel('Scores')
plt.xlabel('Parameter')
lgd = plt.legend(loc=9,bbox_to_anchor=(0.8,1))
plt.show()


print(grid.best_params_)

print('Lasso mean cv R2: {:.4f}'.format(grid.cv_results_['mean_test_score'][para.index(grid.best_params_['alpha'])]))

print("The result improves")
```

```python
para= {'alpha': [1e-3,1e-2, 1, 5, 10,30,50,70,100,500]}
grid = GridSearchCV(Ridge(), para, cv=5)
grid.fit(test, y_train)


para = [1e-3,1e-2, 1, 5, 10,30,50,70,100,500]

plt.figure(1, figsize=(8, 6))

plt.plot(para, grid.cv_results_['mean_test_score'],label='cross-validation score')
plt.plot(para, grid.cv_results_['mean_train_score'],label='training score')

plt.ylabel('Scores')
plt.xlabel('Parameter')
lgd = plt.legend(loc=9,bbox_to_anchor=(0.8,1))
plt.show()


print(grid.best_params_)
print('Ridge mean cv R2: {:.4f}'.format(grid.cv_results_['mean_test_score'][para.index(grid.best_params_['alpha'])]))
print("The result improves just a little")
```

```python
categorical= X_train.dtypes==object
imputer=make_pipeline(SimpleImputer(missing_values=np.nan,strategy='mean'),StandardScaler())

preprocess = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), categorical),
    (imputer,~categorical))

ridge = make_pipeline(preprocess, Ridge(alpha=5))
ridge.fit(X_train,y_train)

coef=ridge.steps[1][1].coef_
ohe_v=ridge.steps[0][1].named_transformers_['onehotencoder'].get_feature_names()

New_column=[]
for i in range (0,len(ohe_v)):
    x,y=ohe_v[i][1:].split('_')
    test=Cat_V[int(x)]+"_"+y
    New_column.append(test)

for each in Non_Cat_v:
    New_column.append(each)

coef_df = pd.DataFrame({'coef': coef,'varname': New_column})

fig, ax = plt.subplots(figsize=(60, 10))


coef_df.plot(x='varname', y='coef', kind='bar', 
            ax=ax, color='none', legend=False)
ax.scatter(x=pd.np.arange(coef_df.shape[0]),s=50,y=coef_df['coef'], color='black')
```

```python
from sklearn.linear_model import Lasso

Lasso = make_pipeline(preprocess, Lasso(alpha=30))
Lasso.fit(X_train,y_train)

coef=ridge.steps[1][1].coef_
ohe_v=ridge.steps[0][1].named_transformers_['onehotencoder'].get_feature_names()

New_column=[]
for i in range (0,len(ohe_v)):
    x,y=ohe_v[i][1:].split('_')
    test=Cat_V[int(x)]+"_"+y
    New_column.append(test)

for each in Non_Cat_v:
    New_column.append(each)

coef_df = pd.DataFrame({'coef': coef,'varname': New_column})
fig, ax = plt.subplots(figsize=(60, 10))

coef_df.plot(x='varname', y='coef', kind='bar', 
            ax=ax, color='none', legend=False)
ax.scatter(x=pd.np.arange(coef_df.shape[0]),s=50,y=coef_df['coef'], color='black')
```

```python
from sklearn.linear_model import ElasticNet

elasticNet = make_pipeline(preprocess, ElasticNet(l1_ratio=1))
elasticNet.fit(X_train,y_train)

coef=ridge.steps[1][1].coef_
ohe_v=ridge.steps[0][1].named_transformers_['onehotencoder'].get_feature_names()

New_column=[]
for i in range (0,len(ohe_v)):
    x,y=ohe_v[i][1:].split('_')
    test=Cat_V[int(x)]+"_"+y
    New_column.append(test)

for each in Non_Cat_v:
    New_column.append(each)

coef_df = pd.DataFrame({'coef': coef,'varname': New_column})

fig, ax = plt.subplots(figsize=(60, 10))


coef_df.plot(x='varname', y='coef', kind='bar', 
            ax=ax, color='none', legend=False)
ax.scatter(x=pd.np.arange(coef_df.shape[0]),s=50,y=coef_df['coef'], color='black')
```
