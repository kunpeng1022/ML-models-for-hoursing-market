```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer, make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import validation_curve
import warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#Task2.1
```

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv',header=0)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"],errors='coerce')
df['TotalCharges'].fillna((df['TotalCharges'].mean()), inplace=True)
df=df.loc[:, df.columns != 'customerID']
```

```python
df_churn=df['Churn'].value_counts()
data = {'No': 5174, 'Yes': 1869}
names = list(data.keys())
values = list(data.values())

fig, axes = plt.subplots(1, 3)
fig.set_size_inches(15, 4)
ax11, ax12, ax13= axes.ravel()

ax11.bar(names, values)
ax11.set_xlabel("Churn")
ax11.set_title("Bar Chart of Churn")

ax12.hist(df['MonthlyCharges'], edgecolor = 'black')
ax12.set_xlabel("MonthlyCharges")
ax12.set_title("Histogram of MonthlyCharges")

ax13.hist(df['TotalCharges'], edgecolor = 'black')
ax13.set_xlabel("TotalCharges")
ax13.set_title("Histogram of TotalCharges")
```

```python
#Task2.2
```

```python
Cat_v=df.loc[:, df.dtypes == object].columns.values
Cat_v=np.delete(Cat_v,15)
Cat_v=np.append(Cat_v,'SeniorCitizen')
C_v=['tenure','TotalCharges','MonthlyCharges']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
X_train, X_test, y_train, y_test = train_test_split(df.loc[:, df.columns != 'Churn'],df['Churn'])
```

```python
preprocess = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), Cat_v))
print("Without Scaler:")
Logistic = make_pipeline(preprocess, LogisticRegression(solver='lbfgs', multi_class='multinomial'))
valid_scores = cross_val_score(Logistic, X_train, y_train,cv=5)
print('LogisticRegression accuracy: {:.4f}'.format(np.mean(valid_scores)))


from sklearn.neighbors import NearestCentroid
Centroid = make_pipeline(preprocess, NearestCentroid())
valid_scores = cross_val_score(Centroid, X_train, y_train,cv=5)
print('NearestCentroid accuracy: {:.4f}'.format(np.mean(valid_scores)))

from sklearn.svm import LinearSVC

LinearSVC = make_pipeline(preprocess, LinearSVC())
valid_scores = cross_val_score(LinearSVC, X_train, y_train,cv=5)
print('LinearSVC accuracy: {:.4f}'.format(np.mean(valid_scores)))
```

```python
#LogisticRegression and LinearSVC accuracy are better
```

```python
preprocess = make_column_transformer((OneHotEncoder(handle_unknown='ignore'), Cat_v),(StandardScaler(),C_v))

print("With Scaler:")

Logistic = make_pipeline(preprocess, LogisticRegression(solver='lbfgs', multi_class='multinomial'))
valid_scores = cross_val_score(Logistic, X_train, y_train,cv=5)
print('LogisticRegression accuracy: {:.4f}'.format(np.mean(valid_scores)))


from sklearn.neighbors import NearestCentroid
Centroid = make_pipeline(preprocess, NearestCentroid())
valid_scores = cross_val_score(Centroid, X_train, y_train,cv=5)
print('NearestCentroid accuracy: {:.4f}'.format(np.mean(valid_scores)))

from sklearn.svm import LinearSVC

SVC = make_pipeline(preprocess, LinearSVC())
valid_scores = cross_val_score(SVC, X_train, y_train,cv=5)
print('LinearSVC accuracy: {:.4f}'.format(np.mean(valid_scores)))
```

```python
#Scaler improves the accuracy for every model.
```

```python
#Task2.3
```

```python
preprocess = make_column_transformer((OneHotEncoder(handle_unknown='ignore'), Cat_v),(StandardScaler(),C_v))
test = preprocess.fit_transform(X_train)

para={'penalty': ['l1','l2'], 'C': [0.001,0.01,0.1,1,5,10]}
grid = GridSearchCV(LogisticRegression(), para, cv=5)
grid.fit(test, y_train)
print(grid.best_params_)

print('LogisticRegression accuracy: {:.4f}'.format(grid.best_score_))

    
plt.figure(1, figsize=(8, 6))

para=[0.001,0.01,0.1,1,5,10]

plt.plot(para, grid.cv_results_['mean_test_score'][::2],label='penalty l1')
plt.plot(para, grid.cv_results_['mean_test_score'][1::2],label='penalty l2')

lgd = plt.legend(loc=9,bbox_to_anchor=(0.85,0.9))
plt.xlabel('C')
plt.ylabel('Validation Score')
plt.show()
```

```python
#The result after gridsearch for logistic regression improves just a little. 
```

```python
para={'C': [0.001,0.01,0.1,1,5,10]}   

grid = GridSearchCV(LinearSVC(), para, cv=5)
grid.fit(test, y_train)

print(grid.best_params_)
print('LinearSVC accuracy: {:.4f}'.format(grid.best_score_))

plt.figure(1, figsize=(8, 6))

para=[0.001,0.01,0.1,1,5,10]

plt.plot(para, grid.cv_results_['mean_test_score'],label='cross-validation score')
plt.plot(para, grid.cv_results_['mean_train_score'],label='training score')
lgd = plt.legend(loc=9,bbox_to_anchor=(0.8,0.8))

plt.xlabel('C')
plt.ylabel('Validation Score')
plt.show()
```

```python
#The result after gridsearch for LinearSVC is nearly the same.
```

```python
para={'metric': ['euclidean','manhattan','cosine']}   


grid = GridSearchCV(NearestCentroid(), para, cv=5)
grid.fit(test, y_train)


print(grid.best_params_)
print('NearestCentroid accuracy: {:.4f}'.format(grid.best_score_))

plt.figure(1, figsize=(8, 6))

para=['euclidean','manhattan','cosine']

plt.plot(para, grid.cv_results_['mean_test_score'],label='cross-validation score')
plt.plot(para, grid.cv_results_['mean_train_score'],label='training score')

lgd = plt.legend(loc=9,bbox_to_anchor=(0.8,0.8))

plt.xlabel('metric')
plt.ylabel('Validation Score')
plt.show()
```

```python
#The result after gridsearch for NearestCentroid is nearly the same.
```

```python
from sklearn.model_selection import KFold
```

```python
print('kfold with shuffling:')
para={'penalty': ['l1','l2'], 'C': [0.001,0.01,0.1,1,5,10]}
grid = GridSearchCV(LogisticRegression(), para, cv=KFold(n_splits=5, random_state=None, shuffle=True))
grid.fit(test, y_train)
print(grid.best_params_)
print('LogisticRegression accuracy: {:.4f}'.format(grid.best_score_))


para={'C': [0.001,0.01,0.1,1,5,10]}   
grid = GridSearchCV(LinearSVC(), para, cv=KFold(n_splits=5, random_state=None, shuffle=True))
grid.fit(test, y_train)
print(grid.best_params_)
print('LinearSVC accuracy: {:.4f}'.format(grid.best_score_))


para={'metric': ['euclidean','manhattan','cosine']}   
grid = GridSearchCV(NearestCentroid(), para, cv=KFold(n_splits=5, random_state=None, shuffle=True))
grid.fit(test, y_train)
print(grid.best_params_)
print('NearestCentroid accuracy: {:.4f}'.format(grid.best_score_))
```

```python
preprocess = make_column_transformer((OneHotEncoder(handle_unknown='ignore'), Cat_v),(StandardScaler(),C_v))
Logistic = make_pipeline(preprocess, LogisticRegression(C=0.1, penalty='l2'))
Logistic.fit(X_train, y_train)
coef=Logistic.steps[1][1].coef_
ohe_v=Logistic.steps[0][1].named_transformers_['onehotencoder'].get_feature_names()

New_column=[]
for i in range (0,len(ohe_v)):
    x,y=ohe_v[i][1:].split('_')
    test=Cat_v[int(x)]+"_"+y
    New_column.append(test)

for each in C_v:
    New_column.append(each)

coef_df = pd.DataFrame({'coef': coef[0],'varname': New_column})

fig, ax = plt.subplots(figsize=(10, 5))

coef_df.plot(x='varname', y='coef', kind='bar', 
            ax=ax, color='none', legend=False)

ax.scatter(x=pd.np.arange(coef_df.shape[0]),s=50,y=coef_df['coef'], color='black')
##this part is from online
```
